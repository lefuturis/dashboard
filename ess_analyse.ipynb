```python
from pyodk.client import Client
import pandas as pd
```

```python
client = Client(config_path="config.toml",cache_path="cache.toml",project_id=2)
client.open()
forms = client.forms.list()
forms
```

```python
cs_json = client.submissions.get_table(form_id="PCMP-100-001-25_HGR")['value']
df = pd.json_normalize(cs_json)
pd.set_option('display.max_columns', None)
df_tanganyika = df[df['dps'] == 'dps2']
df_tanganyika = df_tanganyika[['today','enqueteur','dps', 'zs', 'as','ajout_as','ess','ajout_ess']]
df_grouped = df_tanganyika[['today', 'dps', 'zs']].groupby('zs').size().reset_index(name='Nombre de soumissions')
df_grouped
```
